```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap,LinearSegmentedColormap
plt.rcParams["font.sans-serif"]=["SimHei"] #设置字体
plt.rcParams["axes.unicode_minus"]=False #该语句解决图像中的“-”负号的乱码问题
#数据的加载
```

```python
data = pd.read_excel("./D4到D30数据.xlsx")
data
```

```python
c_list = data.values.tolist()[0]  # 得到想要设置为列索引【表头】的某一行提取出来
data.columns = c_list  		    # 设置列索引【表头】
data.drop([0], inplace=True)
```

```python
data
```

```python
##缺失值填写-1
data1 = data.fillna(-1).values.copy()
data1
```

```python
##填写缺失值是否食用过该食物
j = 0
for i in range(data1.shape[0]):
    for j in range(0,data1.shape[1],5):
        if data1[i,j]>0:
               pass
        elif np.any(data1[i,j+1:j+5]>0):
               data1[i,j] = 1
        else:
               data1[i,j] = 2
```

```python
data1
```

```python
##对于是否使用过该食物进行改正（没有使用过的话就改为2）
j = 0
for i in range(data1.shape[0]):
    for j in range(0,data1.shape[1],5):
        if np.all(data1[i,j+1:j+5]<0):
            data1[i,j] = 2
```

```python
##将使用频率改为每月多少次
j = 0
for i in range(data1.shape[0]):
    for j in range(0,data1.shape[1],5):
        if data1[i,j] == 2:
            data1[i,j+3] =0
            data1[i,j+4] =0
        else:
            if data1[i,j+3]>0:
                pass
            elif data1[i,j+2]>0:
                data1[i,j+3] = 4*data1[i,j+2].copy()
            elif data1[i,j+1]>0:
                data1[i,j+3 ]= 30*data1[i,j+1].copy()
```

```python
data1
```

```python
##将使用频率缺失值用平均值进行填充
for i in range(0,data1.shape[1],5):
    m = np.mean(data1[data1[:,j+3]>0,j+3])
    data1[data1[:,j+3]<0,j+3] =round(m,2)
```

```python
##将平均使用量的缺失值用同频率下的平均值进行填充
for i in range(data1.shape[0]):
    for j in range(0,data1.shape[1],5):
        if data1[i,j+4]<0:
            data1[i,j+4] = round(np.mean(data1[data1[:,j+3] == data1[i,j+3]]),2)
        if data1[i,j+4]<0:
            data1[i,j+4] = round(np.mean(data1[:,j+4]),2)
```

```python
##去掉日周两个频率的列
data2 = data1[:,np.array([0,3,4])+0*5].copy()
for j in range(5,data1.shape[1],5):
    data2 = np.c_["1",data2,data1[:,np.array([0,3,4])+j].copy()]
```

```python
pd.DataFrame(data2).to_excel("./第一问处理之后的文件.xlsx")
```

```python
##统计到底每种食物食用的人数占人数的比例
s = np.array([np.sum(data2[:,j]==1)for j in range(0,data2.shape[1],3)])/data2.shape[0]
s
```

```python
plt.figure(figsize = (20,8),dpi=200)
br = plt.bar(np.arange(s.shape[0]),s)
for i in range(len(br.patches)):
    br.patches[i].set_fc(plt.cm.Accent_r(s[i]))
plt.xticks(np.arange(s.shape[0]),[f"D{i}"for i in range(4,31)])
plt.xlabel("问题编号D4——D30",fontsize = 16)
plt.ylabel("每种食物的食用占总人数的比例",fontsize = 16)
for i in np.arange(s.shape[0]):
    plt.text(i,s[i]+0.02,round(s[i],2),ha = "center")
```

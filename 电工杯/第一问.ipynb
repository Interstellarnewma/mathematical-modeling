```python
import random
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']  # 黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决无法显示符号的问题
palette = 'deep'
sns.set(font='SimHei', font_scale=1.2, palette=palette)  # 解决Seaborn中文显示问题
```

```python
df = pd.read_excel('附件2：调查数据.xlsx')
df = df.iloc[:, 1:]
df = df.dropna()
赋分后df = df.copy()
df
##其中
```

```python
codingDf = pd.read_excel('编码规则.xlsx')
codingDf = codingDf.dropna()
codingDf
```

```python


def findCode(题目, 答案):
    coding = codingDf[(codingDf['题目'] == 题目) & (codingDf['选项'] == 答案)]
    codingScore = int(coding['编码'].values[0])
    return codingScore


findCode('30、您认为人工智能学习工具融合到哪个学习环节？', '评价反馈')
```

```python

def calc(题目, 答案):
    if 答案.find('┋') != -1:
        # 多选题
        答案 = 答案.split('┋')
    else:
        答案 = [答案]
    sum = 0
    for item in 答案:
        sum += findCode(题目, item)
    return sum


calc('24、您认为学习软件与课堂教学相比较最大的优势是什么？', '真题全面┋可以重复学习┋资料全面')
```

```python
for idx, row in df.iterrows():
    print(idx)
    for colIdx, col in enumerate(row):
        赋分后df.iloc[idx, colIdx] = calc(df.columns[colIdx], col)

    
```
